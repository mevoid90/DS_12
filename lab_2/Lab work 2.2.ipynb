```
import random
import math


#генеруэмо тисячу чисел в діапазоні від 0 до 10
list_with_random_numbers = [random.randint(0,100) for i in range(1000)]
                    
def minmax(a):
    mini=min(list_with_random_numbers)
    maxi=max(list_with_random_numbers)
    print("Минимальное число-",mini)
    print("Максимальное число",maxi)
def sr_znach(a):
    sum=0
    for i in range(len(a)):
        sum+=int(a[i])
    sr_arif= sum/len(a)
    
    return sr_arif
def mediana(a):
    input_numbers_sorted=sorted(a)
    print(input_numbers_sorted)
    if (len(a))%2==0:
        s=len(a)//2 
        mediana=(int(input_numbers_sorted[s])+int(input_numbers_sorted[s-1]))/2
    else: mediana=input_numbers_sorted[len(a)//2]
    print('Медиана=',mediana)
def chastota(a):
    dict_text = dict.fromkeys(a) 
    for j in dict_text:
        k=a.count(j)
        dict_text[j]=k      
    sorted_dict = {}
    sorted_keys = sorted(dict_text, key=dict_text.get, reverse=True)  
    for w in sorted_keys:
        sorted_dict[w] = dict_text[w]
    print("Відсортований словарь")
    print(sorted_dict)
def sr_kv_vidh(a):
    
    sum_of_kv=0
    sr_arif=sr_znach(a)
    temp_array=[]
    for i in range(len(a)):
        temp_array.append(a[i]-sr_arif)
    for i in range(len(temp_array)):
        temp_array[i]=temp_array[i]*temp_array[i]
    for i in range(len(temp_array)):
        sum_of_kv+=temp_array[i]
    k=sum_of_kv/(len(a)-1)
    sr_otkl=math.sqrt(k)
    print("Cтандартное отклонение",sr_otkl)


minmax(list_with_random_numbers)
sr_arif=sr_znach(list_with_random_numbers)
print("Среднее арифметическое=",sr_arif)
mediana(list_with_random_numbers)
chastota(list_with_random_numbers)
sr_kv_vidh(list_with_random_numbers)



```
